#### Trabalhando com dados de jogos de tenis
https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv

```python
dados = pd.read_csv(
    "https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv"
)

display(dados.head(5))
```

#### DIversos valores Nulos, textuais, é necessário tratá-los

```python
# Obter informações a respeito de cada coluna
# .info() retorna o tipo dos dados de cada coluna
print(dados.info())
```

#### Descrevendo os dados

```python
display(dados.describe(include='all'))
```

```python
# Verifica a existência de valores nulos
display(dados.isnull())

# Conta quantos nulos há por coluna
print(dados.isnull().sum())
```

#### è possível verificar cada coluna individualmente

```python
display(dados['winner_entry'].describe())

# Retorna os valores únicos da coluna
display(dados['winner_entry'].unique())
```

Iremos apagar todas as linhas onde 'loser_rank' é nulo, pois são pocas ocasiões

```python
dados.dropna(subset=['loser_rank'], inplace=True)
```

Apagar colunas que todos os valores são nulos (se houvesse)

```python
dados.dropna(axis=1)
```

Alterar valor dos dados nulos em uma coluna

```python
print(dados['loser_entry'].unique())

# Substitui por x as linhas da coluna 'loser_entry' onde os valores são null
dados['loser_entry'].fillna(value='x', inplace=True)

print(dados["loser_entry"].unique())
```

Mais atributos

```python
print(dados['w_ace'].unique())

print(dados['loser_ht'].unique)
```

```python
# Substitui os valores NaN nas duas colunas com condições diferentes
dados.fillna({'w_ace': 0, 'loser_ht': dados['loser_ht'].mean}, inplace=True)

print(dados["w_ace"].unique())

print(dados["loser_ht"].unique)
```

Usar o método *ffill* para projetar um dado para o registro seguinte

```python
display(dados['minutes'][1975:1990])

# Método ffill no .fillna() faz com que o registro NaN receba o próximo registro
dados['minutes'].fillna(method='ffill', inplace=True)

display(dados["minutes"][1975:1990])
```

Identificar duplicados

```python
# Mostra se há dados duplicados
display(dados.duplicated())

# Conta quantos dados duplicados existem
print(dados.duplicated().sum())
```

Eliminar duplicados de uma coluna

```python
# drop_duplicates remove todas as linhas duplicadas dado um subset do DF 
dados.drop_duplicates(subset=['tourney_name'], inplace=True)
```

### Fazendo um deslocamento
https://finance.yahoo.com/quote/AAPL/history/?guccounter=1

```python
dados2 = pd.read_csv('./dados/aapl.csv')

display(dados2)
```

```python
# Cria uma coluna e copia o dado dos registros, jogando 1 para baixo, por isso o primeiro registro de close_ontem è NaN, pois não há outro acima dele
dados2['close_ontem'] = dados2['Close'].shift(1)

display(dados2)
```

```python
# Cria uma coluna de porcentagem entre hoje e ontem
dados2['Alteracao %'] = ((dados2['Close'] / dados2['close_ontem'] - 1) * 100)


display(dados2)
```

```python
dados2.dropna(subset=['Alteracao %'], inplace=True)

display(dados2)

display(dados2.describe(include='all'))
```

#### Transformar dados categóricos (qualitativos) em dados numéricos

```python
dados = pd.read_csv(
    "https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv")

display(dados['winner_ioc'].unique())
```

### Usando scikit-learn para preprocessing
* Mapeamento de valores não numéricos em numéricos

```python
from sklearn.preprocessing import LabelEncoder
```

```python
paises = dados['winner_ioc']

print(paises)
```

```python
lencoder = LabelEncoder()

# Transforma os paises em valores numericos, mapeando cada pais para um número
paises_numeros = lencoder.fit_transform(paises)

print(paises_numeros)
```

Atribui os novos números, correspondentes aos países

```python
dados['winner_ioc'] = paises_numeros

display(dados['winner_ioc'])
```
