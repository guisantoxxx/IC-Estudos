### KDD

O kdd possui 5 fases:

* Seleção: que consiste em selecionar um conjunto de dados que farão partes da análise

* Pré-Processamento: consiste em fazer a verificação e limpeza dos dados

* Tranformação: Aplicação de técnicas de transformação (normalização, agregação, criação de novos atributos, redução, etc). Tais ações são feitas sobre atributos úteis nos dados

* Mineração dos Dados: Consiste na aplicação de algoritimos e técnicas para identificar padrões nos dados e verificar hipóteses

* Análise exploratória de Dados: tem como objetivo examinar os dados previamente à aplicação de qualquer técnica estatística. Para realizar a análise exploratória, é necessário conhecer tecnicamente seus dados, o que eles representam e como eles são classificados (IMPORTANTE ENTENDER SEUS DADOS!)

### Demonstração

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Gerar valores aleatórios e plotar um histograma

```python
minutos_por_dia = np.random.randint(0,60,30)

print(minutos_por_dia)

ax = sns.histplot(minutos_por_dia)

ax.set(xlabel='Minutos por Dia', ylabel='Qtd Vezes')

plt.show()
```
