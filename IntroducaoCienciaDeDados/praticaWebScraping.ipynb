#### Importando Bibliotecas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### Fazendo o Request

```python
html = requests.get("https://myanimelist.net/topanime.php").content

soup = BeautifulSoup(html, 'html.parser')
```

### Importando a tabela

```python
tabela = soup.find("table", {"class": "top-ranking-table"}).find_all('tr')

tabela = tabela[1:]

print(tabela)
```

#### Recuperando cada informa챌찾o individual

```python
ranking = []
titulo = []
score = []



for anime in tabela:
    info = anime.findAll("td")
    ranking.append(int(info[0].find("span").text))
    titulo.append(info[1].find("div").find("div", {"class": "di-ib clearfix"}).find('h3').text)
    score.append(float(info[2].find('div').find('span').text))

print(ranking)
print(titulo)
print(score)
```

#### Criando o DF

```python
dados_animes = pd.DataFrame(ranking, columns=['Ranking'])
dados_animes['Titulo'] = titulo
dados_animes['Score'] = score

display(dados_animes)

display(dados_animes[dados_animes['Score'] > 9])
```
