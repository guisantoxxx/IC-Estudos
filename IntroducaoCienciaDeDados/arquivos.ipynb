### Importando as bibliotecas

```python
import pandas as pd
import csv
```

### Lendo um arquivo .txt

```python
# Carrega o arquivo, dado o PATH
arquivo = open('./dados/machado_de_assis.txt', 'r')

# Lê o arquivo, separando as linhas em uma lista
conteudo = arquivo.readlines()

# Fecha o arquivo
arquivo.close()
```

```python
print(conteudo)

for linha in conteudo:
  print(linha)
```

#### Alterando a palavra Rio de Janeiro por Santa isabel

```python
x = ''

for linha in conteudo:
    if linha != '':
        # Concatena x com a linha
        x += linha

print(x)

print('-----')

x = x.replace('Rio de Janeiro', 'Santa Isabel')

print(x)
```

#### Gravando, (Gerar um novo arquivo)

```python
# abre um arquivo no modo escrita
arquivo_novo = open('./dados/troca_cidade.txt', 'w')

arquivo_novo.writelines(x)

arquivo_novo.close()
```

### Trabalhando com arquivos CSV

#### Abrindo o arquivo.csv

```python
with open('./dados/TSLA.csv', 'r') as file:
  # separados do arquivo é vírgula
  leitor = csv.reader(file, delimiter=',')

  data = []
  fechamento = []
  volume = []

  for linha in leitor:
    data.append(linha[0])
    fechamento.append(linha[4])
    volume.append(linha[5])

# para remover os titulos
data = data[1:]
fechamento = fechamento[1:]
volume = volume[1:]

dados = pd.DataFrame(data, columns=['Data'])
dados['Fechamento'] = fechamento
dados['Volume'] = volume

display(dados)
```

#### Tratando Arquivos usando o Pandas

Pegando direto do GitHub

```python
# armazena o link do arquivo
arquivo = "https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv"

#Função que carrega o arquivo dado o link ou o PATH
dados2 = pd.read_csv(arquivo)

display(dados2)
```

#### Plotando uma infomação

```python
# Ainda é possível fazer operações sobre os dados, como contar e ordenar. Por fim, ainda é possível plotar
# Plotando como barra horizontal
# Nesse caso, seleciona a mão do jogador vencedor, faz a contagem de lados iguais, ordena e depois plota
print(dados2.winner_hand.value_counts().sort_values()
      .plot(kind='barh'))
```

#### Abrir o arquivo da Tesla usando o Pandas

```python
# Skip rows ignora as primeiras x linhas, nesse caso, está pulando a primeira linha para remover os nomes das colunas
# Depois nomeia todas as colunas pelo parametro names
dados = pd.read_csv('./dados/TSLA.csv', skiprows=1, names=['Data', 'Abertura', 'Alta', 'Baixa', 'Fechamento', 'Fechamento Adjunto', 'Volume'], nrows=4)

display(dados)
```
